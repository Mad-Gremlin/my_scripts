```python
entire = "/home/sryaxovsky/Daphnia_pulex_direct_go.txt"
GO_dict = {}
GO_list = []
results_list = {}
parthenogenetic_results_list = {}
counter = 1
annotation2 = []
with open(entire) as en:
    for line in en:
        line = line.split()
        GO_dict[line[0]] = line[1]
for line in GO_dict.values():
    line = line.split("|")
    GO_list.append(line)
GO_list = [g[0] for g in GO_list] 
```

```python
from itertools import groupby
```

```python
### открваем аннотацию eggnog и вытаскиваем все GO термины
annotation = "/mnt/data/lizards/dasha/val/emapp_pep_val.emapper.annotations"
with open(annotation) as an:
    for contig in an:
        contig = contig.split()
        annotation2.append(contig)
            # gene in GO_list:
               #results_list[contig[0]] += 1
       #    elif gene not in GO_list:
        #       results_list[contig[0]] = 0
for i in annotation2:
    if i[0] == "#" or i[0] == "#query_name":
        annotation2.remove(i)
    else:
        continue

final_GO = {}
for n in annotation2:
    final_GO[n[0]] = n[5]   ## получаем файл final_GO со всеми контигами и их GO терминами
```

```python
## закидываем все GO термины в список test_list
test_list = []
for key, values in final_GO.items():
    values = values.split(",")
    test_list.append(values)
    

```

```python
### объединяем группу списков один список 
flat_test_list = [item for sublist in test_list for item in sublist]
```

```python
## фильтруем список, избавляясь от не GO терминов и убираем дупликаты
Vse_go = []
for i in flat_test_list:
    if i.startswith("GO"):
        #rint(i)
        Vse_go.append(i)  
    else:
        continue
        
Vse_go = [el for el, _ in groupby(Vse_go)]
```

```python
with open ("go_list", "w") as fh:
    for i in Vse_go:
        fh.write("%s\n" %(i))
#print(*Vse_go, sep="\n")
```
